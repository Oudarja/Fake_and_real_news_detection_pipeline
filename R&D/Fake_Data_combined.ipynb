```python
df=pd.read_csv("LabeledFake-1K.csv", encoding="utf-8")
```

```python
# Ensure content is string and handle missing values
df["content"] = df["content"].fillna("").astype(str)

# Create title: first sentence (or first line)
df["title"] = df["content"].str.split(r"[।\n]").str[0].str.strip()

# Create description: remaining text
df["description"] = df["content"].apply(
    lambda x: " ".join(x.split("।")[1:]).strip()
)

# Drop original content column
df.drop(columns=["content"], inplace=True)

# Reorder columns
df = df[["title", "description", "label"]]

# Save to CSV
df.to_csv("LabeledFake-1K_.csv", index=False, encoding="utf-8")

print(df.head())
```
